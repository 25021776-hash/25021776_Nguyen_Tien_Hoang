```python
#W6A20
def search(a,x):
    for i in range(len(a)):
       if a[i] == x:
           return i
    return -1

a = list(map(int,input().split()))
x = int(input())
print(search(a,x))
```

```python
#W6A21
a = list(map(str, input().split()))
d ={}
for i in range (len(a)):
    if  a[i] not in d.keys():
        d[a[i]] = [i]
    else:
        d[a[i]].append(i)
m = {}
for k, v in d.items():
    m[k] = tuple(v)
print(m)
```

```python
#W6A23
def erase(a,x):
    b = list(set(a))
    for i in b:
        if i < x:
            b.remove(i)
    return list(b)
a = list(map(int,input().split()))
x = int(input())
print(erase(a,x))
```

```python
#W6A26
a = list(map(int, input().split()))
sum = 0
for i in range(1,len(a) - 1):
    if a[i] > a[i + 1] and a[i] > a[i - 1]:
        sum += a[i]
print(sum)
```
