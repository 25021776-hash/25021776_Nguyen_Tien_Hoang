```python
#W5A1
def lon(a,b):
    if a > b:
        return a
    else:
        return b
if __name__ == '__main__':
    a,b = map(int,input().split())
    print(lon(a,b))
```

```python
#W5A2
c =[]
def swap(a,b):
    a , b = b, a
    c.append(a)
    c.append(b)
    return tuple((c))
a, b = map(int, input().split())
print(swap(a,b))
```

```python
#W5A3
from math import *
def prime(n):
    if n < 2:
        return False
    for i in range(2, isqrt(n)):
        if n % i == 0:
            return False
    return True
n = int(input())
print(prime(n))
```

```python
#W5A4
def hoanhao(n):
    sum = 0
    for i in range(1,n):
        if n % i == 0:
            sum += i
    if sum == n:
        return True
n = int(input())
print(hoanhao(n))
```

```python
#W5A6
def gt(n):
    m = 1
    if n == 0:
        return 1
    for i in range(1,n + 1):
        m *= i
    return m
k = int(input())
print(gt(k))
```

```python
#W5A7
def operator(a,b,c):
    if c == "+":
        return a + b
    elif c == "-":
        return a - b
    elif c == "*":
        return a * b
    elif c == "/":
        return a / b
    else:
        return False
a,b = map(int,input().split())
c = input()
print(operator(a,b,c))
```

```python
#W5A8
def binary(a,c):
    if a == 0:
        c.append(0)
    elif a == 1:
        c.append(1)
    else:
        c.append(a % 2)
        return binary(a // 2,c)
    return c

a, b = map(int, input().split())
c = []
d = []
binary(a,c)
binary(b,d)
c.reverse()
d.reverse()
m = max(len(c),len(d))
cnt = 0
while len(c) < m:
    c.insert(0,0)
while len(d) < m:
    c.insert(0,0)
for i in range(m):
    if c[i] != d[i]:
        cnt += 1
print(cnt)
```

```python
#W5A9
def tong(n):
    sum = 0
    while n != 0:
        sum += n % 10
        n = n // 10
    return sum
n = int(input())
print(tong(n))
```
