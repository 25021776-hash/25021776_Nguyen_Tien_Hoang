```python
#W6A9
m,n = map(int, input().split())
a = {}
b = {}
c = {}
for i in range(m):
    k,v = input().split()
    a[k] = int(v)
for i in range(n):
    k,v = input().split()
    b[k] = int(v)
for k,v in a.items():
    if k in b:
        c[k] = b[k] + v
    else:
        c[k] = a[k]
for k,v in b.items():
    if k not in c:
        c[k] = v
sorted_keys = sorted(c.keys())
sorted_c = {k: c[k] for k in sorted_keys}
print(sorted_c)
```

```python
#W6A10
a = list(map(int, input().split()))
k = int(input())
b =[]
for i in range(len(a)):
    for j in range(i + 1, len(a)):
        c = []
        if a[i] + a[j] == k:
            c.append(a[i])
            c.append(a[j])
        if len(c) != 0:
            b.append(tuple(c))
print(b)
```

```python
#W6A12
a= list(map(int, input().split()))
cnt = [0] * (1000000)
for i in a:
    cnt[i] += 1
ans = 0
res = 0
for i in range(1000000):
    if cnt[i] > ans:
        ans = cnt[i]
        res = i
print(res,ans)
```
